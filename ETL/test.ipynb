```python
import pandas as pd 
import os
```

```python
data = pd.read_csv("../Prepared Data/customer.csv")
data.head(1)
```

```python
customer_deal = data.groupby(by="customer_id", as_index=False).agg(deals=("customer_id", "count"))
customer_deal
```

```python
customer_deal = customer_deal.groupby(by="deals", as_index=False).count()
customer_deal
```

```python
folder_path = '../Prepared Data'
if not os.path.exists(folder_path):
    os.makedirs(folder_path)

# Save each DataFrame to a CSV file in the new folder
customer_deal.to_csv(os.path.join(folder_path, 'customer_deal.csv'), index=False)
```
